```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
num_elements = 20
```

### Stationary without Noise

```python
stationary_without_noise = np.full(num_elements, 5)
plt.plot(stationary_without_noise)
plt.xlabel("Timestep"), plt.ylabel("Value"), plt.title("Stationary without Noise")
```

### Stationary with Gaussian Noise

```python
stationary_with_gaussian_noise = np.random.normal(5, 0.1, num_elements)
plt.plot(stationary_with_gaussian_noise)
plt.xlabel("Timestep"), plt.ylabel("Value"), plt.title("Stationary with Gaussian Noise")
```

### Stationary with Non-Gaussian Noise

```python
stationary_with_non_gaussian_noise = np.random.standard_t(3, num_elements) + 5
plt.plot(stationary_with_non_gaussian_noise)
plt.xlabel("Timestep"), plt.ylabel("Value"), plt.title("Stationary with Non-Gaussian Noise")
```

### Linear Trending without Noise

```python
linear_trending_without_noise = (np.arange(num_elements) / num_elements - 0.5) * 2 + 5
plt.plot(linear_trending_without_noise)
plt.xlabel("Timestep"), plt.ylabel("Value"), plt.title("Linear Trending without Noise")
```

### Linear Trending with Non-Gaussian Noise

```python
linear_trending_with_non_gaussian_noise = (np.arange(num_elements) / num_elements - 0.5) * 2 + 5 + np.random.standard_t(3, num_elements)
plt.plot(linear_trending_with_non_gaussian_noise)
plt.xlabel("Timestep"), plt.ylabel("Value"), plt.title("Linear Trending with Non-Gaussian Noise")
```

```python
num_features = 5

# generate correlation matrices
amplitude_corr = np.full((num_features, num_features), 1)
frequency_corr = np.full((num_features, num_features), 1)
phase_corr = np.full((num_features, num_features), 1)

# Generate correlated parameters
amplitude_means = np.random.uniform(1, 5, num_features)
frequency_means = np.random.uniform(0.001, 0.01, num_features)
phase_means = np.random.uniform(0, 2 * np.pi, num_features)

# Generate correlated deviations from means
amplitude_deviations = np.random.normal(scale=np.sqrt(1 - amplitude_corr), size=(num_features, num_features))
frequency_deviations = np.random.normal(scale=np.sqrt(1 - frequency_corr), size=(num_features, num_features))
phase_deviations = np.random.normal(scale=np.sqrt(1 - phase_corr), size=(num_features, num_features))

# Apply Cholesky decomposition to ensure positive definite covariance matrix
amplitude_cov = amplitude_corr * np.eye(num_features) + amplitude_deviations @ amplitude_deviations.T
frequency_cov = frequency_corr * np.eye(num_features) + frequency_deviations @ frequency_deviations.T
phase_cov = phase_corr * np.eye(num_features) + phase_deviations @ phase_deviations.T

amplitude_params = np.random.multivariate_normal(mean=amplitude_means, cov=amplitude_cov, size=num_elements)
frequency_params = np.random.multivariate_normal(mean=frequency_means, cov=frequency_cov, size=num_elements)
phase_params = np.random.multivariate_normal(mean=phase_means, cov=phase_cov, size=num_elements)

time_series = np.zeros((num_elements, num_features))
time = np.arange(num_elements)
for i in range(num_features):
  time_series[:, i] = amplitude_params[:, i] * np.sin(2 * np.pi * frequency_params[:, i] * time + phase_params[:, i])
```

```python
# number of timeseries
num_features = 5

# correlation matrix between timeseries
start_range, end_range = 0.8, 1
lower_triangular = np.tril(np.random.uniform(start_range, end_range, size=(num_features, num_features)), k=-1)
corr = lower_triangular + lower_triangular.T
np.fill_diagonal(corr, 1)

# Set random seed for reproducibility
np.random.seed(42)

# Define parameters
start_date = '2020-01-01'
end_date = '2023-12-31'
mean_return = np.full(num_features, 0.0005)  # Mean daily return
volatility = 0.02  # Daily volatility

# Generate dates
dates = pd.date_range(start=start_date, end=end_date, freq='B')
num_periods = len(dates)

# Generate random returns based on a normal distribution
returns = np.random.multivariate_normal(mean=mean_return, cov=corr, size=num_periods) * volatility

# Calculate cumulative returns
cumulative_returns = np.exp(np.cumsum(returns, axis=0))

# Set initial price
initial_price = 100

# Calculate prices
prices = initial_price * cumulative_returns

print(dates.shape)
print(prices.shape)

# Create DataFrame
data = {'Date': dates}
for i, price_timeseries in enumerate(prices.T):
  data[i] = price_timeseries

df = pd.DataFrame(data)

# Plot the time series
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
for col in df.columns:
  if col != 'Date':
    plt.plot(df['Date'], df[col], label=col)
plt.title('Synthetic Financial Time Series')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True)
plt.legend()
plt.show()
```
